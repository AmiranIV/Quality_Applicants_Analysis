```python
import pandas as pd


# Group by 'Applicant' ID and calculate the mean of grades
data_mean = data.groupby('Applicant').mean().reset_index()

# Print the transformed dataset
data_mean.head()
```

```python
new_data = data_mean.drop(data_mean.columns[1],axis=1)
new_data.head()
```

```python
new_data['final_result'] = new_data.mean(axis=1)
```

```python
new_data.head()
```
